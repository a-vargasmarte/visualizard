```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import re
from difflib import SequenceMatcher
from datetime import datetime
%matplotlib inline
```

```python
data = pd.read_csv('regionMesSubrupo')
```

```python
data.head()
```

```python
vizSunburst= data[['Region', 'Mes', 'Sub-Grupo de servicios/ PDSS', 'cantidadServiciosTotal', 'poblacionTotal', 'valorTotal', 'gastoPerCapitaTotal', 'valorPct', 'gastoPerCapitaPct']]
```

```python
vizSunburst.head()
```

```python
timesSunburst = vizSunburst.Mes.str.split(expand=True)
```

```python
for index, row in timesSunburst.iterrows():
    if row[0] == 'Enero':
        row[0] = 'january'
    elif row[0] == 'Febrero':
        row[0] = 'february'
    elif row[0] == 'Marzo':
        row[0] = 'march'
    elif row[0] == 'Abril':
        row[0] = 'april'
    elif row[0] == 'Mayo':
        row[0] = 'may'
    elif row[0] == 'Junio':
        row[0] = 'june'
    elif row[0] == 'Julio':
        row[0] = 'july'
    elif row[0] == 'Agosto':
        row[0] = 'august'
    elif row[0] == 'Septiembre':
        row[0] = 'september'
    elif row[0] == 'Octubre':
        row[0] = 'october'
    elif row[0] == 'Noviembre':
        row[0] = 'november'
    elif row[0] == 'Diciembre':
        row[0] = 'december'
    
    
```

```python
timesSunburst.head()
```

```python
for index, row in timesSunburst.iterrows():
    row[0] = '1 ' + row[0] + ', ' + row[1]
```

```python
timesSunburst.head()
```

```python
s = timesSunburst[0][0]
d = datetime.strptime(s, '%d %B, %Y')
print(d.strftime('%Y-%m-%d'))
```

```python
for index, row in timesSunburst.iterrows():
    s = row[0]
    d = datetime.strptime(s, '%d %B, %Y')
    row[0] = d.strftime('%Y-%m-%d')
```

```python
vizSunburst['Mes'] = timesSunburst[0]
```

```python
vizSunburst.rename(columns={'Mes': 'time', 'Sub-Grupo de servicios/ PDSS': 'subgrupo' }, inplace=True)
```
