```python
def municipiosArray():
    municipioArray = []
    for municipio in geodata['features']:
        name = municipio['properties']['NAME_2']
        municipioArray.append(name)
    return municipioArray
```

```python
def lowerCaseMunicipios():
    lowerMunicipio = []
    for municipio in uniqueMunicipios:
#         print(municipio)
        test = re.sub(r'\([^)]*\)', '', municipio).lower()
        secondFilter = re.sub(r'd.m', '', test)
        thirdFilter = re.sub(r'dm', '', secondFilter)
#         print(secondFilter)
        lowerMunicipio.append(thirdFilter)
        
    return lowerMunicipio
        
```

```python
municipiosClean = lowerCaseMunicipios()
municipiosClean
```

```python
geoLowerMunicipio = []
for geoMunicipio in geoMunicipios:
    geoMunicipio = geoMunicipio.lower()
    geoLowerMunicipio.append(geoMunicipio)

municipiosClean.sort()

geoLowerMunicipio.sort()

print(SequenceMatcher(a=municipiosClean[1], b=geoLowerMunicipio[0]).ratio())
```

```python
geoMunicipioClean = []
municipioClean = []
for municipio in municipiosClean:
    for geoMunicipio in geoLowerMunicipio:
        if SequenceMatcher(a=municipio, b=geoMunicipio).ratio() > 0.85:
#             print(geoMunicipio)
            municipio = geoMunicipio
            geoMunicipioClean.append(geoMunicipio)
            municipioClean.append(municipio)
```

```python
def filterMunicipio(x):
    test = re.sub(r'\([^)]*\)', '', x).lower()
    secondFilter = re.sub(r'd.m', '', test)
    thirdFilter = re.sub(r'dm', '', secondFilter)
    return thirdFilter
```

```python
def correctGramar(y):
    for x in geolower:
        if SequenceMatcher(a=x, b=y).ratio() > 0.88:
#             print(x, y, SequenceMatcher(a=x, b=y).ratio())
            return x
        else:
            print(y)
```
